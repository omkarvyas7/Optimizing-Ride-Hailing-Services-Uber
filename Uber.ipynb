<h2>Matching in Ride-Hailing</h2>
<br>
<font size="+1">
    <ul>
        <li>Rides can be either non-shared, meaning that the ride is arranged for only one customer group, or shared, meaning that several customer groups with different pickup and dropoff locations share the ride.</li>
        <br>
        <li>In the simpler case of a non-shared ride, drivers participating on the platform cycle through three states sequentially: </li>
        <br>
        <ul>
            <li><i>"open"</i> (waiting to be dispatched),</li>
            <br>
            <li><i>"en route"</i> (on the way to the pickup location),</li>
            <br>
            <li>and <i>"on-trip"</i> (driving riders to their destination).</li>
            <br>
        </ul>
        <li>A request can be matched with a dispatchable driver using very simple algorithms, such as the <b><i>first-dispatch protocol</i></b>.</li>
        <br>
        <ul>
            <li>In the first-dispatch protocol only open drivers are considered as dispatchable.</li>
            <br>
            <li>Each request is immediately assigned to the open driver who is predicted to have the shortest en route time.</li>
            <br>
            <li>This matching algorithm is also called the <i>on-call</i> policy, the <i>closest driver</i> policy, and the <i>committed driver-rider</i> matching protocol.</li>
            <br>
        </ul>
        <li>An alternative matching approach is <i>batching</i> where rider requests are collected for a short time window (typically on the order of seconds), at the end of which an optimization problem is solved to pair each request with an open driver.</li>
        <br>
        <li>If there are riders that are not matched in this batch, then they are carried over and used for the optimization problem in the next batching window.</li>
        <br>
        <li>Batching can reduce the rider waiting time relative to the <i>first-dispatch</i> protocol by consolidating requests and making better use of supply.</li>
        <br>
        <ul>
            <li>Note that the first-dispatch protocol can be viewed as a special case of the batching protocol when each batch consists of exactly one rider request.</li>
            <br>
        </ul>
        <li>Due to its substantial benefits, batching has been implemented widely in major ride-hailing platforms.</li>
        <br>
    </ul>
</font>

<h2>Implemented the First-Dispatch Routing Protocol with a Maximum Dispatch Radius</h2>

```python
# Find a single driver with a minimum ETA
# Find a single driver with minimum ETA that satisfies the MDR (maximum dispatch radius) constraint
```

```python
import numpy as np
import pandas as pd

import cvxpy as cp
```

```python
rng = np.random.default_rng(42)
```

```python
# Number of Riders
num_riders = 1

# Number of drivers
num_drivers = rng.poisson(lam=20)

# Goal is to route DRIVER to RIDER
# Equivalently, to match a rider with a driver

# Physical measure of distance between riders and drivers
distances = pd.DataFrame(rng.uniform(0, 10, (num_riders, num_drivers)),
                         columns=[f'Driver {j}' for j in range(1, num_drivers+1)],
                         index=[f'Rider {i}' for i in range(1,num_riders+1)])

# Map distances to a binary constraint which consists 1 if within MDR and 0 otherwise
# A value of 1 means the driver and rider can be connected.
MDR = 5 # don't dispatch outside of MDR miles
maximum_dispatch_radius = (distances < MDR).apply(lambda x: np.where(x, 1, 0)) # np.where(distances < MDR, 1, 0)

# Temporal measure of distance between riders and drivers
# Expected Time of Arrival
# ETA should be a function of the distance, driver's allowable speed, and some idiosyncratic noise
# (need a more realistic way to assign a variance, maybe as a function of distance)
eta = distances.apply(lambda x: rng.pareto(x))# rng.exponential(x)) 
```

```python
eta
```

```python
# distances = distances.transform(lambda x: x+20) # Sanity Check

distances
```

```python
maximum_dispatch_radius
```

```python
# Decision Variables
```

<p>Decision variable is a binalry/boolean variable of size num_drivers that maintains whether a driver is chosen or not for the rider</p>
<p>Taking boolean varibale as the decision is whether to select a driver or not i.e. 0 or 1 </p>

```python
all_drivers = cp.Variable(num_drivers, boolean = True)
```

```python
# Objective
```

<p> The objective of the problem is to minimize ETA of the rider-driver pair </p>



$$\begin{align}
\textbf{Minimize :} \\
& \text{for all drivers i} \sum_{i}{\text{all_drivers}_{i}\times \text{eta}_{i}}
\end{align}$$

```python
#Looping through each driver
for i in range(0, eta.shape[1]):
        
        #First value therefore assign
        if i == 0:
            objective_eta = all_drivers[i] * eta.iloc[0,i]
        
        #Subsequent value therefore add
        else:
            objective_eta += (all_drivers[i] * eta.iloc[0,i])

#Objective Function to minimize

objective_function = cp.Minimize(objective_eta)
```

```python
# Constraint
```

<p> The constraint to this problem are as follows : 
    <ol>
        <li><b>MDR Constraint :</b> Only drivers within MDR should be considered i.e. where MDR = 1 </li>
        <li><b>Single Driver Constraint :</b> Only 1 driver should be choosen</li>
    </ol>
</p>  

$$\begin{align}
\textbf{Constraint to :} \\
&\text{MDR}_{driver} = 1 \\
& \sum_{i}{\text{all_drivers}_{i}} = 1 \\
\end{align}$$

```python
#Extracting MDS Matrix to numpy array
driver_mdr = maximum_dispatch_radius.values[0]

#Variable to maintain constraints
all_constraints = []

#Looping through MDR array to apply constraint
for i in range(0, len(driver_mdr)):
    
    #Checking if MDR for particular driver = 0 then do not choose that driver
    if driver_mdr[i] == 0:
        
        #Assinging 0 in the decision variable for such drivers
        all_constraints.append((all_drivers[i] == 0))

#Adding constraint that only 1 driver should be selected
all_constraints.append(sum(all_drivers) == 1)
```

```python
# Solution in terms of a dispatching recommendation
```

```python
#Problem Formulation
prob = cp.Problem(objective_function, all_constraints)

#Obtaining Optimal Solution i.e. the minimum eta for the rider-driver assignment
optimal_solution = prob.solve()
```

```python
#Converting the driver assgnment array to dataframe
first_dispatch_protocol_suggestion = pd.DataFrame(all_drivers.value.astype(int)).T

#Assigning column name for clear representation
first_dispatch_protocol_suggestion.columns = eta.columns

#Assigning index for clear representation
first_dispatch_protocol_suggestion.index = eta.index

first_dispatch_protocol_suggestion
```

```python
#Printing the First-Dispatch Protocol Suggestion
print('------------------------ First-Dispatch Protocol Suggestion ------------------------')

#Looping through the final decision dataframe
for i in range(0, len(first_dispatch_protocol_suggestion.T)):
    
    #Checking which driver is selected 
    if first_dispatch_protocol_suggestion.iloc[0,i] == 1:
        print('\nRider 1 ----- Driver {}\
                \nMatch Rider 1 with Driver {} to obtain a optmal/minimum ETA = {}'\
              .format(i + 1, i + 1, np.round(optimal_solution,4)))
```

<h2>Batching Dispatch Protocol: Minimized the Aggregated En Route Time Model</h2>

```python
rng = np.random.default_rng(42)
```

```python
# Number of Riders
num_riders = rng.poisson(lam=15)

# Number of drivers
num_drivers = rng.poisson(lam=15)

# Goal is to route DRIVER to RIDER
# Equivalently, to match a rider with a driver

# Physical measure of distance between riders and drivers
distances = pd.DataFrame(rng.uniform(0, 30, (num_riders, num_drivers)),
                         columns=[f'Driver {j}' for j in range(1, num_drivers+1)],
                         index=[f'Rider {i}' for i in range(1,num_riders+1)])

# Temporal measure of distance between riders and drivers
# Expected Time of Arrival
# ETA should be a function of the distance, driver's allowable speed, and some idiosyncratic noise
# (need a more realistic way to assign a variance, maybe as a function of distance)
eta = distances.apply(lambda x: rng.pareto(x)) 
```

```python
eta
```

```python
distances
```

```python
# Decision Variables
```

<p>Decision variable is a binary/boolean variable of size (num_riders, num_drivers) that maintains whether a driver-rider pair is chosen or not </p>
<p>Taking boolean varibale as the decision is whether to select a driver-rider pair or not i.e. 0 or 1 </p>

```python
rider_driver = cp.Variable((num_riders, num_drivers), boolean = True)
```

```python
# Objective
```

<p> The objective of the problem is to minimize ETA of the overall rider-driver pairs </p>



$$\begin{align}
\textbf{Minimize :} \\
& \text{for all rider i and driver j} \sum_{i,j}{\text{all_drivers}_{i,j}\times \text{eta}_{i,j}}
\end{align}$$

```python
#Looping through the eta dataframe to calculate objective sum
for i in range(0,rider_driver.shape[0]):
    
    for j in range(0,rider_driver.shape[1]):
        
        #First value therefore assign
        if j == 0 and i == 0:
            objective_eta = (rider_driver[i,j] * eta.iloc[i,j])
        
        #Subsequent value therefore add
        else:
            objective_eta += (rider_driver[i,j] * eta.iloc[i,j])
```

```python
#Minimize
objective_function = cp.Minimize(objective_eta)
```

```python
# Constraints
```

<p> The constraint to this problem are as follows : 
    <ol>
        <li><b>Rider Constraint :</b> 
            <ul>
                <li>When num_drivers = num_riders : Rider should be assigned one driver</li>
                <li>When num_drivers > num_riders : Rider should be assigned one driver</li>
                <li>When num_riders > num_drivers : Rider can be assigned zero or one driver</li>
            </ul>
        </li><br>
        <li><b>Driver Constraint :</b> 
            <ul>
                <li>When num_drivers = num_riders : Driver should be assigned one rider</li>
                <li>When num_drivers > num_riders : Driver can be assigned zero or one rider</li>
                <li>When num_riders > num_drivers : Driver should be assigned one rider</li>
            </ul>
        </li>
    </ol>
</p>  

$$\begin{align}
\textbf{Constraint to :} \\
& \text{When num_drivers = num_riders} \\
&& \text{for all rider i } \sum_{j}{\text{driver_rider}_{i,j}} = 1 \\
&& \text{for all driver i } \sum_{j}{\text{driver_rider}_{j,i}} = 1 \\
& \text{When num_drivers > num_riders} \\
&& \text{for all rider i } \sum_{j}{\text{driver_rider}_{i,j}} = 1 \\
&& \text{for all driver i } \sum_{j}{\text{driver_rider}_{j,i}} \leq 1 \\
& \text{When num_drivers < num_riders} \\
&& \text{for all rider i } \sum_{j}{\text{driver_rider}_{i,j}} \leq 1 \\
&& \text{for all driver i } \sum_{j}{\text{driver_rider}_{j,i}} = 1 \\
\end{align}$$

```python
#Combined constraint list
all_constraints = []

#Rider Sum i.e. row sum
for i in range(0, rider_driver.shape[0]):
    for j in range(0, rider_driver.shape[1]):
        
        #First value therefore assign
        if j == 0:
            rider_sum = rider_driver[i,j]
        
        #First value therefore add
        else:
            rider_sum += rider_driver[i,j]
    
    #Case 1 : num_drivers == num_riders : constraint rider sum = 1
    if num_drivers == num_riders:
        all_constraints.append((rider_sum == 1))
    
    #Case 2 : num_drivers > num_riders : constraint rider sum = 1
    elif num_drivers > num_riders:
        all_constraints.append((rider_sum == 1))
    
    #Case 3 : num_drivers < num_riders : constraint rider sum <= 1
    else:
        all_constraints.append((rider_sum <= 1))

#Driver Sum i.e column sum
for i in range(0, rider_driver.shape[1]):
    for j in range(0, rider_driver.shape[0]):
        
        #First value therefore assign
        if j == 0:
            driver_sum = rider_driver[j,i]
            
        #First value therefore add
        else:
            driver_sum += rider_driver[j,i]
    
    #Case 1 : num_drivers == num_riders : constraint driver sum = 1
    if num_drivers == num_riders:
        all_constraints.append((driver_sum == 1))
    
    #Case 2 : num_drivers > num_riders : constraint driver sum = 1
    elif num_drivers > num_riders:
        all_constraints.append((driver_sum <= 1))
    
    #Case 3 : num_drivers < num_riders : constraint driver sum <= 1
    else:
        all_constraints.append((driver_sum == 1))  

all_constraints
```

```python
# Solution in terms of a dispatching recommendation
```

```python
#Problem Formulation
prob = cp.Problem(objective_function, all_constraints)

#Obtaining Optimal Solution i.e. the minimum eta for combined rider-driver pairs assignment
optimal_solution = prob.solve()
```

```python
#Converting the driver-rider pair assignment array to dataframe
batch_protocol_suggestion = pd.DataFrame(
                                        rider_driver.value.astype(int),
                                        columns = eta.columns,
                                        index = eta.index
                                        )
batch_protocol_suggestion
```

```python
#Printing the Batch Protocol Suggestion
print('------------------------ Batch Protocol Suggestion ------------------------ \n')

#Looping through the final decision dataframe
for i in range(0, batch_protocol_suggestion.shape[0]):
    for j in range(0, batch_protocol_suggestion.shape[1]):
        
        #Checking which driver-rider pair is selected 
        if batch_protocol_suggestion.iloc[i,j] == 1:
            print('Rider {} ----- Driver {}'.format(i + 1, j + 1))

print('\nThe above suggestion from the Batch Protocol will lead to a optimal/minimum batch ETA = {}\n'\
     .format(np.round(optimal_solution,4))
     )

#Checking if all Riders were matched
all_rider = 0

#Check for Riders
for i in range(0, batch_protocol_suggestion.shape[0]):
    rider_count = 0
    for j in range(0, batch_protocol_suggestion.shape[1]):
        
        rider_count += batch_protocol_suggestion.iloc[i,j]
    
    if rider_count == 0:
        print('Rider {} was not matched. Will be carried over to next batch'.format(i + 1))
        all_rider += 1

#Check if all riders were matched
if all_rider == 0:
    print('All Riders were matched')
        
```

<h2>Batching Dispatch Protocol: Maximized Profit Model with a Maximum Dispatch Radius</h2>

```python
rng = np.random.default_rng(42)
```

```python
# Number of Riders
num_riders = rng.poisson(lam=15)

# Number of drivers
num_drivers = rng.poisson(lam=15)

# Goal is to route DRIVER to RIDER
# Equivalently, to match a rider with a driver

# Physical measure of distance between riders and drivers
distances = pd.DataFrame(rng.uniform(0, 30, (num_riders, num_drivers)),
                         columns=[f'Driver {j}' for j in range(1, num_drivers+1)],
                         index=[f'Rider {i}' for i in range(1,num_riders+1)])

# Map distances to a binary constraint which consists 1 if within MDR and 0 otherwise
# A value of 1 means the driver and rider can be connected.
MDR = 15 # don't dispatch outside of MDR miles
maximum_dispatch_radius = (distances < MDR).apply(lambda x: np.where(x, 1, 0)) # np.where(distances < MDR, 1, 0)

# Temporal measure of distance between riders and drivers
# Expected Time of Arrival
# ETA should be a function of the distance, driver's allowable speed, and some idiosyncratic noise
# need a more realistic way to assign a variance, maybe as a function of distance
profits = distances.apply(lambda x: x + rng.lognormal(mean=1, sigma=3)) 
```

```python
# Decision Variables
```

<p>Decision variable is a binalry/boolean variable of size (num_riders, num_drivers) that maintains whether a driver rider is chosen or not</p>
<p>Taking boolean varibale as the decision is whether to select a driver-rider pair or not i.e. 0 or 1 </p>

```python
rider_driver = cp.Variable((num_riders, num_drivers), boolean = True)
```

```python
# Objective
```

<p> The objective of the problem is to Maximize Profit of the overall rider-driver pairs </p>



$$\begin{align}
\textbf{Maximize :} \\
& \text{for all rider i and driver j} \sum_{i,j}{\text{rider_driver}_{i,j}\times \text{profits}_{i,j}}
\end{align}$$

```python
#Looping through the eta dataframe to calculate objective sum
for i in range(0,rider_driver.shape[0]):
    
    for j in range(0,rider_driver.shape[1]):
        
        #First value therefore assign
        if j == 0 and i == 0:
            objective_profit = (rider_driver[i,j] * profits.iloc[i,j])
        
        #Subsequent value therefore add
        else:
            objective_profit += (rider_driver[i,j] * profits.iloc[i,j])
```

```python
#Maximize
objective_function = cp.Maximize(objective_profit)
```

```python
# Constraints
```

<p> The constraint to this problem are as follows : 
    <ol>
        <li><b>Rider Constraint :</b> 
            <ul>
                <li>When num_drivers = num_riders : Rider should be assigned one driver</li>
                <li>When num_drivers > num_riders : Rider should be assigned one driver</li>
                <li>When num_riders > num_drivers : Rider can be assigned zero or one driver</li>
            </ul>
        </li><br>
        <li><b>Driver Constraint :</b> 
            <ul>
                <li>When num_drivers = num_riders : Driver should be assigned one rider</li>
                <li>When num_drivers > num_riders : Driver can be assigned zero or one rider</li>
                <li>When num_riders > num_drivers : Driver should be assigned one rider</li>
            </ul>
        </li><br>
        <li><b>MDR Constraint : </b>Only drivers within MDR should be considered i.e. where MDR = 1 
        </li>
    </ol>
</p>  

$$\begin{align}
\textbf{Constraint to :} \\
&\text{for all rider i and driver j, MDR}_{i,j} = 1 \\
\\
& \text{When num_drivers = num_riders} \\
&& \text{for all rider i } \sum_{j}{\text{driver_rider}_{i,j}} = 1 \\
&& \text{for all driver i } \sum_{j}{\text{driver_rider}_{j,i}} = 1 \\
& \text{When num_drivers > num_riders} \\
&& \text{for all rider i } \sum_{j}{\text{driver_rider}_{i,j}} = 1 \\
&& \text{for all driver i } \sum_{j}{\text{driver_rider}_{j,i}} \leq 1 \\
& \text{When num_drivers < num_riders} \\
&& \text{for all rider i } \sum_{j}{\text{driver_rider}_{i,j}} \leq 1 \\
&& \text{for all driver i } \sum_{j}{\text{driver_rider}_{j,i}} = 1 \\
\end{align}$$

```python
all_constraints = []

#Rider Sum i.e. row sum
for i in range(0, rider_driver.shape[0]):
    for j in range(0, rider_driver.shape[1]):
        
        #First value therefore assign
        if j == 0:
            rider_sum = rider_driver[i,j]
        
        #First value therefore add
        else:
            rider_sum += rider_driver[i,j]
    
    #Case 1 : num_drivers == num_riders : constraint rider sum = 1
    if num_drivers == num_riders:
        all_constraints.append((rider_sum == 1))
    
    #Case 2 : num_drivers > num_riders : constraint rider sum = 1
    elif num_drivers > num_riders:
        all_constraints.append((rider_sum == 1))
    
    #Case 3 : num_drivers < num_riders : constraint rider sum <= 1
    else:
        all_constraints.append((rider_sum <= 1))

#Driver Sum i.e column sum
for i in range(0, rider_driver.shape[1]):
    for j in range(0, rider_driver.shape[0]):
        
        #First value therefore assign
        if j == 0:
            driver_sum = rider_driver[j,i]
            
        #First value therefore add
        else:
            driver_sum += rider_driver[j,i]
    
    #Case 1 : num_drivers == num_riders : constraint driver sum = 1
    if num_drivers == num_riders:
        all_constraints.append((driver_sum == 1))
    
    #Case 2 : num_drivers > num_riders : constraint driver sum = 1
    elif num_drivers > num_riders:
        all_constraints.append((driver_sum <= 1))
    
    #Case 3 : num_drivers < num_riders : constraint driver sum <= 1
    else:
        all_constraints.append((driver_sum == 1))  
    
        
#MDR Constraint
for i in range(0, rider_driver.shape[0]):
    
    for j in range(0, rider_driver.shape[1]):
        
        #Checking if the driver-rider pair falls outside the MDR limit 
        if maximum_dispatch_radius.iloc[i,j] == 0:
            
            #Assinging 0 in the decision variable for such driver-rider pair
            all_constraints.append((rider_driver[i,j] == 0))

all_constraints
```

```python
# Decision Variables
```

<p>Decision variable is a binary/boolean variable of size (num_notifications, time_periods) that maintains whether to send a notification at the particular time slot or not.</p>
<p>Taking boolean varibale as the decision is whether to send a notification at the particular time slot or not i.e. 0 or 1 </p>

```python
push_notification = cp.Variable((num_notifications, time_periods), boolean = True)
```

```python
# Objective
```

<p>The objective of this problem is to Maximize the Score for the Push Notification and Time Slot pairs</p>

$$\begin{align}
\textbf{Maximize :} \\
& \text{for all notification i and time slot t} \sum_{i,t}{\text{push_notification}_{i,t}\times \text{scores}_{i,t}}
\end{align}$$

```python
#Looping through the scores to calculate the objective score
for i in range(0, scores.shape[0]):
    
    for j in range(0, scores.shape[1]):
        
        #First value therefore assign
        if j == 0 and i == 0:
            objective_score = (push_notification[i,j] * scores.iloc[i,j])
        
        #Subsequent value therefore add
        else:
            objective_score += (push_notification[i,j] * scores.iloc[i,j]) 
```

```python
#Maximize
objective_function = cp.Maximize(objective_score)
```

```python
# Constraints
```

<p> The constraint to this problem are as follows : 
    <ol>
        <li><b>Count for each Push Notification Constraint :</b> The maximum number of time a single/particular push notification can be sent (in the current scenario it is defined as 1 but can be changed if required).</li><br>
        <li><b>Count for each Time Slot Constraint :</b> The maximum number of notifications that can be sent in a particular time slot (in the current scenario it is defined as 2 but can be changed if required).</li><br>
        <li><b>Total Notification Constraint</b> The maximum total number of notifications that can be sent combined across all time slots (in the current scenario it is defined as 5 but can be changed if required).</li>
    </ol><br>
<b>The motivation behind choosing the constraints :</b>

As this is open-ended, I have assumed each time slot to be a day and the constaints have been configured based on this. There will have cases when at any particular time there will be multiple push notifications that can extract good potential but of we restrict that to only 1 notification in the time slot we might lose out on values derived. But this does not mean that we send 10 push in a time slot because that tends to make user unlikely to follow up on because it feels like you have been bombarded with push notifications. 2 Push notifications at any time slot seems appropriate where we can have push notification from same or different departments of the company eg. 1 discounts/offers push notification and 1 new restaurants push notification. Uber Eats does not want to lose users by sending big amount of push notifications. 

Moreover each push notification should only be sent 1 time because the next push notification can be further customised based on the user activity between the time slot that particualr notification was last send and current time slot. So sending the same push notification might not be the most optimal way. 

Also, restricting the total number of notifications sent across time slots to 5 serves the purpose of again not sending big number of notifications to the users. People tend to treat such cases as spams and do get a sense of personalisation. We can instead use user activity data from these 5 push notification and further customise and personlise the future push notification.

<b>Note :</b> All the numbers can be controlled through the variable defined in following cells. 
</p>  

$$\begin{align}
\textbf{Constraint to :} \\
& \text{for each push notification i } \sum_{t}{\text{push_notification}_{i,t}} \leq 1 \\
& \text{for each time slot t } \sum_{i}{\text{push_notification}_{i,t}} \leq 2 \\
& \text{for all push notification i and time slot t } \sum_{i,t}{\text{push_notification}_{i,t}} \leq 5 \\
\end{align}$$

```python
all_constraints = []

#Control for number of times a notification can be pushed in the time horizon
num_push_for_notfication_i = 1

#Control for number of notifications that can be pushed in the time slot t
num_notfications_for_time_slot_t = 2

#Control for Maximum number of notifications that can be pushed in the entire time horizon
num_maximum_notfications_for_time_horizon = 5

#1. Push notification i at most num_push_for_notfication_i times

for i in range(0, scores.shape[0]):
    for j in range(0, scores.shape[1]):
        
        if j == 0:
            notification_i_count = push_notification[i,j]
            
        else:
            notification_i_count += push_notification[i,j]
    
    all_constraints.append((notification_i_count <= num_push_for_notfication_i))
    
#2. Send at most num_notfication_for_time_slot_t Push notifications at time slot t

for i in range(0, scores.shape[1]):
    for j in range(0, scores.shape[0]):
        
        if j == 0:
            time_t_count = push_notification[j,i]
            
        else:
            time_t_count += push_notification[j,i]
    
    all_constraints.append((time_t_count <= num_notfications_for_time_slot_t))
    
#3. Maximum number of notifications that can be pushed in the entire time horizon

for i in range(0, scores.shape[0]):
    for j in range(0, scores.shape[1]):
        
        if j == 0 and i == 0:
            total_notification_count = push_notification[i,j]
            
        else:
            total_notification_count += push_notification[i,j]

#Adding constraint for total number of notification allowed across all time slots
all_constraints.append((total_notification_count <= num_maximum_notfications_for_time_horizon))


#Note : Constraint that Push notification i at time t or don't is controlled by taking the variable as boolean type \
# and controlling for above constraints 1 & 2

all_constraints
```

```python
# Solution in terms of a scheduling recommendation for the push notifications
```

```python
#Problem Formulation
prob = cp.Problem(objective_function, all_constraints)

#Optimal solution i.e. the push notification and time slot pair that gives maximum score
optimal_solution = prob.solve()
```

```python
#Converting final decision array to dataframe 
push_notification_scheduling = pd.DataFrame(
                                        push_notification.value.astype(int),
                                        columns = scores.columns,
                                        index = scores.index
                                        )
push_notification_scheduling 
```

```python
#Printing Push Notification Schedule 
print('------------------------ Push Notification Schedule ------------------------ \n')

#Looping through the decision dataframe
for i in range(0, push_notification_scheduling.shape[1]):
    for j in range(0, push_notification_scheduling.shape[0]):
        
        #Checking if the particular noteification and time slot pair is selected
        if push_notification_scheduling.iloc[j,i] == 1:
            print('Time Slot {} ----- Push Notification {}'.format(i + 1, j + 1))

print('\nThe above Push Notification Scheduling will lead to a optimal/maximum Score = {}'\
     .format(np.round(optimal_solution,2))
     )
    
```
